```python
#this program uses the heuristic function as a guide to find the solution to our problem, which is how to reach the goal state of the puzzle from the initial state
#here the various states of the board on shifting the space are the various nodes of the tree and we go to the node with the lowest heuristic
#keep 2 lists one closed to look at nodes already explored and their neighbours already discovered
```

```python
class Node: #this tells me what the current state of the puzzle is and functions to move the empty space to create new neghboring classes
    def __init__(self,data,depth,fscore):
        self.data = data
        self.depth = depth
        self.fscore = fscore

    def generate_child(self): #func to create the child node by moving the tile thats empty or 0
        x,y = self.find(self.data,'0')
        val_list = [[x,y-1],[x,y+1],[x-1,y],[x+1,y]]
        children = []
        for i in val_list:
            child = self.shuffle(self.data,x,y,i[0],i[1])
            if child is not None:
                child_node = Node(child,self.depth+1,0)
                children.append(child_node)
        return children
        
    def shuffle(self,puz,x1,y1,x2,y2): #function that shifts the space part if its possible to shift it 
        if x2 >= 0 and x2 < len(self.data) and y2 >= 0 and y2 < len(self.data):
            temp_puz = []
            temp_puz = self.copy(puz)
            temp = temp_puz[x2][y2]
            temp_puz[x2][y2] = temp_puz[x1][y1]
            temp_puz[x1][y1] = temp
            return temp_puz
        else:
            return None
            

    def copy(self,root): #this is to create a copy of the puzzle
        temp = []
        for i in root:
            t = []
            for j in i:
                t.append(j)
            temp.append(t)
        return temp    
            
    def find(self,puz,x): #function that tries to see where the peice marked x is 
        for i in range(0,len(self.data)):
            for j in range(0,len(self.data)):
                if puz[i][j] == x:
                    return i,j


class Puzzle: #this takes in the goal and initial state and calculates the heuristic
    def __init__(self):
        self.n = 3
        self.open = []
        self.closed = []

    def accept(self): #function to input the initial state of the puzzle
        puz = []
        for i in range(0,self.n):
            temp = input().split(" ")
            puz.append(temp)
        return puz

    def f(self,start,goal): #heuristic function
        return self.h(start.data,goal)+start.depth

    def h(self,start,goal): #manhattan distance
        temp = 0;
        for i in range(0,self.n):
          for j in range(0, self.n):
            if goal[i][j]!=start[i][j]:
              for p in range(0,self.n):
                for q in range(0,self.n):
                  if goal[p][q]==start[i][j]:
                    temp=temp+abs((i-p)+(j-q))
                    break
        return temp
    
    def create(self):  #take in the end and start state
        print("Enter the start state\n")
        start = self.accept()
        print("Enter the goal \n")        
        goal = self.accept()
        start = Node(start,0,0)
        start.fscore = self.f(start,goal)
        self.open.append(start) #appending the open list
        print("\n\n")
        while True:
            current_state = self.open[0]
            print("\n")
            for i in current_state.data:
                for j in i:
                    print(j,end=" ")
                print("")
            if(self.h(current_state.data,goal) == 0):
                break
            for i in current_state.generate_child():
                i.fscore = self.f(i,goal)
                self.open.append(i)
            self.closed.append(current_state)
            del self.open[0]
            self.open.sort(key = lambda x:x.fscore,reverse=False)
puz = Puzzle()
puz.create()
```
